```python
import numpy as np 

import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
 df=pd.read_csv('H:\\SixflagellaCheYp.csv')
x=df.iloc[:,0]
F1 = df.iloc[:,1] 
F2 = df.iloc[:,2] 
F3 = df.iloc[:,3] 
F4 = df.iloc[:,4] 
F5 = df.iloc[:,5] 
F6 = df.iloc[:,6] 
```

```python

__author__ = 'vafadar'

__version__ = '1.0'

plt.rcParams.update({'font.family': 'Arial'})
plt.rcParams.update({'font.size': 30})

       
fig, axs = plt.subplots(2, 3, sharey=True,figsize=(15,10))

#fig = plt.figure(figsize=(10,10))

#ax = fig.add_subplot(111)

col='orange'

axs[0, 0].scatter(x, F1, lw=4, color =col)
axs[0, 1].scatter(x, F2, lw=4, color =col)
axs[0, 2].scatter(x, F3, lw=4, color =col)

axs[1, 0].scatter(x, F4, lw=4, color =col)
axs[1, 1].scatter(x, F5, lw=4, color =col)
axs[1, 2].scatter(x, F6, lw=4, color=col)


for i in [0,1]:
    for j in [0,1, 2]:
        axs[i,j].tick_params(axis='x', which='both', bottom=True, top=False,
            labelbottom=True, size=20, width=3, direction='out', pad=10, colors='black')
        axs[i,j].tick_params(axis='y', which='both', left=True, right=False,
            labelbottom=True, size=20, width=3, direction='out', pad=10, colors='black')
        axs[i,j].spines['right'].set_visible(True)
        axs[i,j].spines['top'].set_visible(True)
        axs[i,j].spines['bottom'].set_visible(True)
        axs[i,j].spines['left'].set_visible(True)
        axs[i,j].spines['left'].set_linewidth(3)
        axs[i,j].spines['bottom'].set_linewidth(3)
        axs[i,j].spines['right'].set_linewidth(3)
        axs[i,j].spines['top'].set_linewidth(3)
        axs[i,j].spines['bottom'].set_color('black')
        axs[i,j].spines['left'].set_color('black')
        if i == 0:
            axs[i,j].set_xticklabels(['', ''])
        axs[i,j].set_xticks([0, 50, 100]) 
#start, end = axs.get_xlim()

#plt.xticks(np.arange(0, 100, 50))

plt.savefig('h:\\SixflagellaCheYp.png', bbox_inches='tight', dpi=300)

plt.show()
```
