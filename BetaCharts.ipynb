```python
import numpy as np 
from scipy.stats import beta
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x=  np.arange(-60., 60., 1)
rv1 = beta(1, 1)
rv2 = beta(2, 2)
rv3 = beta(3, 3)
rv4 = beta(4, 4)
rv5 = beta(5, 5)
rv6 = beta(6, 6)
rv7 = beta(7, 7)
rv8 = beta(8, 8)
rv9 = beta(9, 9)
rv10 = beta(10, 10)
```

```python


__author__ = 'vafadar'

__version__ = '1.0'


fig = plt.figure(figsize=(10,10))
plt.rcParams.update({'font.family': 'Arial'})
plt.rcParams.update({'font.size': 30})
ax = fig.add_subplot(111)
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')

# put your x and y arrays or lists here.        

ax.errorbar(x,rv1.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=1$')
ax.errorbar(x,rv2.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=2$')
ax.errorbar(x,rv3.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=3$')
ax.errorbar(x,rv4.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=4$')
ax.errorbar(x,rv5.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=5$')
ax.errorbar(x,rv6.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=6$')
ax.errorbar(x,rv7.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=7$')
ax.errorbar(x,rv8.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=8$')
ax.errorbar(x,rv9.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=9$')
ax.errorbar(x,rv10.pdf((x+60)/120) , lw=5, label=r'$\alpha = \beta=10$')

#fig = plt.figure()
#ax = fig.add_subplot(111)
#ax.plot(x,a,'k--',x,b,'k:',x,c,'k')


handles, labels = ax.get_legend_handles_labels()

ax.legend(handles,labels, loc='best', fontsize='small', \
          handlelength=0.5, handletextpad=0.4, framealpha=0., labelspacing=0.1,\
          numpoints=1, markerscale=0.3,  fancybox=False, edgecolor='white')

plt.tick_params(axis='x', which='both', bottom=True, top=False,
    labelbottom=True, size=20, width=3, direction='out', pad=25, colors='black')
plt.tick_params(axis='y', which='both', left=True, right=False,
    labelbottom=True, size=20, width=3, direction='out', pad=25, colors='black')



#plt.tight_layout()




plt.savefig('h:\\beta.png', bbox_inches='tight', dpi=300)
plt.show()
```
